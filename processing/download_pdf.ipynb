```python
packages = ['pymongo', 'gridfs']
for package in packages:
    try:
        __import__(package)
    except ImportError:
        !pip install {package}

import os
from pymongo import MongoClient
from functions.pdf_operations import PDFOperations

#endpoint = '10.0.10.5:27017' # vpn
endpoint = '192.168.1.10:27017'
database = 'supercias'
collection = 'companies'
pdfs_path = '../storage/pdfs'

# Connect to MongoDB
db = MongoClient('mongodb://'+endpoint)[database]
# Create instance of PDFOperations
pdf_ops = PDFOperations(db, collection, pdfs_path)
# Get all companies
companies = db[collection]
```

### Make a query to get the pdfs we are interested in for a given company

```python
companie = companies.find_one({'ruc': '0991295437001'})
docs = companie['Documentos']

# pdfs filenames 
pdfs_filenames = []

#if company has documents
for value in docs.values():
    if isinstance(value, list) and len(value) > 0:
        [ pdfs_filenames.append(v) for v in value ]

print(pdfs_filenames)
```
