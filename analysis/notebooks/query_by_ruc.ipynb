# Query the company by RUC
This script will connect to the mongodb and query the database by ruc and the consulta personal for that ruc

```python
from pymongo import MongoClient

endpoint = '10.0.10.5:27017'
database = 'supercias'
companies_collection = 'ranking'
cp_collection = 'consulta_personal'

# Connect to MongoDB
db = MongoClient('mongodb://'+endpoint)[database]
cp_collection = db[cp_collection]
companies_collection = db[companies_collection]

#find one person
person = cp_collection.find_one()
name = person['nombre']
print(name)

# find one company
company = companies_collection.find_one()
print(company['nombre'])

```

## Query compania by ruc

```python
def query_company_by_ruc(ruc):
    company = companies_collection.find_one({'ruc': ruc})
    # remove _id field
    if(company):
        del company['_id']
    return company
```

## Query persona personal by ruc

```python
def query_person_by_ruc(ruc):
    # query the person
    query = {
        "$or": [
            { "accionista_actual.Ruc": ruc },  # Searching in accionista_actual
            { "accionista_anteriores.Ruc": ruc },  # Searching in accionista_anteriores
            { "administrador_actual.Ruc": ruc },  # Searching in administrador_actual
            { "administradores_anteriores.Ruc": ruc },  # Searching in administradores_anteriores
            { "accionista_extrajeras.Ruc": ruc }  # Searching in accionista_extrajeras
        ]
    }
    # find the persons
    persons = cp_collection.find(query)
    # conver to list
    persons = list(persons)
    # remove the _id field
    for person in persons:
        del person['_id']
    return persons

```

### With a given Ruc find all of the values comapies and people in 

```python
ruc = '0992524200001'
company = query_company_by_ruc(ruc)
people = query_person_by_ruc(ruc)
print({'company': company, 'people': people})
```

```python
# ruc to query``
ruc = '1891715664001'
# open json file to write to 
def query_by_ruc(ruc):
    company = query_company_by_ruc(ruc)
    people = query_person_by_ruc(ruc)
    return {'company': company, 'people': people}

print(query_by_ruc(ruc))
```
